```
from pydub import AudioSegment
def clip(path, t1, t2):
  t1 = t1 * 1000 #Works in milliseconds
  t2 = t2 * 1000
  newAudio = AudioSegment.from_wav(path)
  newAudio = newAudio[t1:t2]
  newAudio.export('/content/newSong.wav', format="wav")
```

```
def graph(path):
  #import the pyplot and wavfile modules 
  import matplotlib.pyplot as plot
  from scipy.io import wavfile
  
  # Read the wav file (mono)
  samplingFrequency, signalData = wavfile.read(path)
  
  # Plot the signal read from wav file
  plot.subplot(211)
  plot.title('Spectrogram of %s' %path)
  #plot.xscale('symlog', linthreshx = .5)
  plot.plot(signalData)
  plot.xlabel('Sample')
  plot.ylabel('Amplitude')

  plot.subplot(212)
  plot.specgram(signalData,Fs=samplingFrequency)
  plot.xlabel('Time')
  plot.ylabel('Frequency')
  plot.show()
```

```
graph('/content/audio1_i_word.wav')
```
