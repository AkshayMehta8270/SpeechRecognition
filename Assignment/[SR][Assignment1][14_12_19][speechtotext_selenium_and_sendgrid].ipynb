# Selenium function to open URL

```python
def open_url(s):
    driver = webdriver.Chrome('chromedriver.exe')
    driver.get(s)
    time.sleep(5)
    driver.close()
```

# Regex to check if Email or URL

```python
import regex
def email_or_site(s):
    
    s_email = regex.search(r'[\w\.-]+@[\w\.-]+', s)
    if s_email is not None:
        s_email = s_email.group(0)
        print('Email: ', s_email)
        
    #If email is found, then call email function
    if s_email is not None:
        s = s_email
        send_mail(s)
        
    s_url = regex.search(r'\b(https|http|)(://)?(www\.)?\w+\.(com|org)\b', s)
    if s_url is not None:
        s_url = s_url.group(0)
        print('URL: ', s_url)
    
    #If URL is found, then call URL function
    if s_url is not None:
        print(s_url)
        #Adding 'https://'' if not there
        if s_url[0:3] == 'www':
            s_url = 'https://' + s_url
        elif s_url[0:5] == 'https':
            pass
        #Adding 'https://www.'' if not there
        else: 
            s_url = 'https://www.' + s_url
        s = s_url
        open_url(s)
```
